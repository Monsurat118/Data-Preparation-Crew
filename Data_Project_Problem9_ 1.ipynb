Replace 0 in colum Number of employees

```python
def fill_with_mode_replace_zeros(x):
    mode_val = x[x != 0].mode()  # Compute mode excluding 0s
    median_val = x[x != 0].median()  # Compute median excluding 0s
    mean_val = x[x != 0].mean()  # Compute mean excluding 0s

    if not mode_val.empty:
        x = x.replace(0, mode_val.iloc[0])  # Replace 0s with mode
    elif not pd.isna(median_val):
        x = x.replace(0, median_val)  # If no mode, use median
    elif not pd.isna(mean_val):
        x = x.replace(0, mean_val)  # If no median, use mean
    else:
        x = x.replace(0, 1)  # Last fallback: Replace with 1 (or another predefined value)

    return x

df_clean['Number of employees'] = df_clean.groupby([naics_col, province_col, company_col, year_col])['Number of employees'].transform(fill_with_mode_replace_zeros)
```

```python


# Define columns to clean (handling outliers)
columns_to_clean = [
    "Off-site Disposal - Landfill",
    "Off-site Transfers for Recycling - Energy Recovery",
    "Off-site Transfers for Recycling - Other",
    "Off-site Transfers for Recycling - Recovery of Acids and Bases",
    "Off-site Transfers for Recycling - Recovery of Catalysts",
    "Off-site Transfers for Recycling - Recovery of Inorganic Materials (not metals)",
    "Off-site Transfers for Recycling - Recovery of Metals and Metal Compounds",
    "Off-site Transfers for Recycling - Recovery of Organic Substances (not solvents)",
    "Off-site Transfers for Recycling - Recovery of Pollution Abatement Residues",
    "Off-site Transfers for Recycling - Recovery of Solvents",
    "Off-site Transfers for Recycling - Refining or Re-use of Used Oil",
    "Off-site Transfers for Treatment Prior to Final Disposal - Biological Treatment",
    "Off-site Transfers for Treatment Prior to Final Disposal - Chemical Treatment",
    "Off-site Transfers for Treatment Prior to Final Disposal - Incineration / Thermal",
    "Off-site Transfers for Treatment Prior to Final Disposal - Municipal Sewage Treatment Plant",
    "Off-site Transfers for Treatment Prior to Final Disposal - Physical Treatment",
    "On-site Disposal - Landfill"
]

# Define grouping columns (Fixed the issue)
grouping_columns = ["Reporting_Year / Année", "Substance Name (English) / Nom de substance (Anglais)"]

# Ensure year column is in integer format
df_clean[grouping_columns[0]] = df_clean[grouping_columns[0]].astype(int)

# Function to replace outliers with median per group
def replace_outliers_with_median(df, columns, group_cols):
    total_outliers_replaced = 0  # Counter for total outliers replaced

    for column in columns:
        if column in df.columns:  # Check if the column exists in the dataset
            # Compute IQR per group
            grouped = df.groupby(group_cols)
            Q1 = grouped[column].quantile(0.25)
            Q3 = grouped[column].quantile(0.75)
            IQR = Q3 - Q1

            # Define outlier bounds
            lower_bounds = Q1 - 1.5 * IQR
            upper_bounds = Q3 + 1.5 * IQR

            # Compute median per group
            medians = grouped[column].median()

            # Track number of outliers replaced
            outliers_count = 0

            # Replace outliers with the median of that group
            def replace(row):
                nonlocal outliers_count
                group_values = tuple(row[col] for col in group_cols)
                value = row[column]

                if group_values in lower_bounds.index and pd.notna(value):
                    lower = lower_bounds[group_values]
                    upper = upper_bounds[group_values]
                    median = medians[group_values]

                    if value < lower or value > upper:
                        outliers_count += 1
                        return median  # Replace with median of the same group

                return value

            df[column] = df.apply(replace, axis=1)
            total_outliers_replaced += outliers_count

            print(f"Total outliers replaced in '{column}': {outliers_count}")

    return df, total_outliers_replaced

# Apply the function to clean outliers across multiple columns
df_clean, num_outliers_replaced = replace_outliers_with_median(df_clean, columns_to_clean, grouping_columns)

# Print total outliers replaced across all columns
print(f"\nTotal outliers replaced across all columns: {num_outliers_replaced}")
```

```python
# Define shorter column names
short_names = {
    "Off-site Disposal - Landfill": "Landfill Disp.",
    "Off-site Transfers for Recycling - Energy Recovery": "Energy Rec.",
    "Off-site Transfers for Recycling - Other": "Recycling Other",
    "Off-site Transfers for Recycling - Recovery of Acids and Bases": "Acids/Bases Rec.",
    "Off-site Transfers for Recycling - Recovery of Catalysts": "Catalysts Rec.",
    "Off-site Transfers for Treatment Prior to Final Disposal - Biological Treatment": "Bio Treatment",
    "Off-site Transfers for Treatment Prior to Final Disposal - Chemical Treatment": "Chem Treatment",
    "Off-site Transfers for Treatment Prior to Final Disposal - Incineration / Thermal": "Incineration",
    "Off-site Transfers for Treatment Prior to Final Disposal - Municipal Sewage Treatment Plant": "Sewage Treatment",
    "Off-site Transfers for Treatment Prior to Final Disposal - Physical Treatment": "Phys Treatment",
}

# Select the desired columns and rename them
columns_for_heatmap = list(short_names.keys())
heatmap_df = df_clean_columns[columns_for_heatmap].rename(columns=short_names)

# Calculate the correlation matrix
correlation_matrix = heatmap_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")

# Rotate labels for better readability
plt.xticks(rotation=45, ha="right")
plt.title("Correlation Heatmap of Releases")
plt.show()
```
